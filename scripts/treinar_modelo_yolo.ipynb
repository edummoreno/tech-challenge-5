```python
# Importar bibliotecas
from ultralytics import YOLO
import os
import shutil
```

```python
# Montar o drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Copiar o dataset + data.yaml file do drive para o content do colab
shutil.copytree(
    "/content/drive/MyDrive/Tech Challenge 5/yolo_dataset",
    "/content/yolo_dataset",
    dirs_exist_ok=True
)
```

```python
# Carregar o modelo YOLO
last_model_path = "/content/drive/MyDrive/Tech Challenge 5/resultados_yolo/base_train/weights/last.pt"

if os.path.exists(last_model_path):
    print("Modelo encontrado: carregando last.pt")
    model = YOLO(last_model_path)
else:
    print("last.pt não encontrado: carregando yolov8m.pt")
    model = YOLO("yolov8m.pt")
```

```python
# Configurar a flag da variavel resume do modelo yolo
resume_flg = None

if os.path.exists(last_model_path):
    print("Modelo continuará o treinamento a partir do last.pt")
    resume_flg = True
else:
    print("Modelo começará o treinamento do zero")
    resume_flg = False
```

```python
# Treinar o modelo yolo
yolo_results = model.train(
    data='/content/yolo_dataset/data.yaml',
    epochs=50,
    batch=14,
    imgsz=768,
    workers=4,
    resume=resume_flg,
    single_cls=False,
    device=0, # GPU
    amp=True,
    project='/content/drive/MyDrive/Tech Challenge 5/resultados_yolo',
    name='base_train',
    box=7.5,
    cls=1.0,
    dfl=1.5,
    val=True,
    patience=10,
    degrees=0.0,
    hsv_s=0.0,
    hsv_v=0.0,
    scale=0.2,
    fliplr=0.0,
    cache='disk'
)
```

```python
# Carregar o best.pt do modelo yolo treinado
best_model_path = str(yolo_results.save_dir / "weights" / "best.pt")

trained_model = YOLO(best_model_path)
```

```python
# Aplicar FINE TUNING ao modelo treinado
fine_tuning_results = trained_model.train(
    data='/content/yolo_dataset/data.yaml',
    epochs=20,
    batch=12,
    imgsz=832,
    workers=4,
    resume=False,
    single_cls=False,
    device=0, # GPU
    amp=True,
    project='/content/drive/MyDrive/Tech Challenge 5/resultados_yolo',
    name='fine_tuning_train',
    box=7.5,
    cls=1.0,
    dfl=1.5,
    val=True,
    patience=8,
    degrees=0.0,
    hsv_s=0.0,
    hsv_v=0.0,
    scale=0.2,
    fliplr=0.0,
    cache='disk'
)
```
