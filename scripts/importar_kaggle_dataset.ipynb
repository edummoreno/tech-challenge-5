```python
# Importar as bibliotecas
import os, shutil, kagglehub
```

```python
# Montar o google drive
from google.colab import drive
drive.mount("/content/drive")
```

```python
# Carregar o arquivo kaggle.json
os.makedirs('/root/.kaggle', exist_ok=True)

shutil.copy(
    '/content/drive/MyDrive/Tech Challenge 5/kaggle.json',
    '/root/.kaggle/kaggle.json'
)

os.chmod('/root/.kaggle/kaggle.json', 0o600)

print("kaggle.json configurado")
```

```python
# Fazer o download do dataset via kagglehub
dataset_ref = "carlosrian/software-architecture-dataset"
path = kagglehub.dataset_download(dataset_ref)

print("Dataset baixado.")
```

```python
# Copiar o dataset para dentro do drive
DEST_DIR = "/content/drive/MyDrive/Tech Challenge 5/dataset"

# Verifica se a pasta de destino ja existe e depois copia os arquivos para ela
if os.path.exists(DEST_DIR):
    print(f"Pasta destino já existe: {DEST_DIR}")
    print("Será criada uma cópia com sufixo _copy para não sobrescrever.")
    DEST_DIR = DEST_DIR + "_copy"

print(f"Copiando para: {DEST_DIR}")
shutil.copytree(path, DEST_DIR)

print("Cópia concluída!")
```
